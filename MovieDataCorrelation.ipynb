```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)
```

```python
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int)
```

```python
#change data type of columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
#create correct year column TEST
df['correct year'] = df['released'].str.extract(r'([1-2]\d{3})')
df
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
df.corr() # default is pearson correlation
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for all Numeric Features')
```

```python
# Practice with numerizing data types. These data should not be accounted for in any 
# real analysis. Categories such as "director", "writer", "star", and "company" do not have meaningful
# values when converted to numerals.
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
plt.subplots(figsize=(15, 15))
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for all Numeric Features After Numerizing the DataFrame')
```

```python
correlation_matrix = df_numerized.corr()
corr_pairs = correlation_matrix.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

### There is a meaningful correlation between the 'gross' and 'votes' columns and the 'gross' and 'budget' columns.

```python

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```
